### Feature Engineering

```python
# Menghitung usia properti berdasarkan tahun pembangunan
df['property_age'] = 2025 - df['made']

# Menjumlahkan total fasilitas tambahan (basement, attic, garage)
df['total_special_rooms'] = df['basement'] + df['attic'] + df['garage']

# Menghitung harga per meter persegi
df['price_per_sqm'] = df['price'] / df['squareMeters']
```

### Outlier Checking

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
num_cols
```

```python
# Check Outlier using IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    
    return outliers.shape[0]

for col in num_cols:
    print(col, ":", detect_outliers_iqr(df, col))
```

**Note:**

  Terdapat outlier pada kolom baru `price_per_sqm` dengan jumlah 1264

```python
"""
Capping (atau disebut juga winsorizing) adalah teknik menangani outlier dengan cara:

1. Mengganti nilai ekstrem dengan batas maksimum atau minimum tertentu
2. BUKAN menghapus datanya

Jadi datanya tetap ada, hanya nilainya yang “dipotong” ke batas wajar.
"""

def cap_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    data[column] = data[column].clip(lower, upper)
    
for col in num_cols:
    cap_outliers(df, col)
```

```python
for col in num_cols:
    print(col, ":", detect_outliers_iqr(df, col))
```
